```python
import pm4py
from pm4py.algo.evaluation.generalization import algorithm as generalization_evaluator
from pm4py.algo.evaluation.simplicity import algorithm as simplicity_evaluator
from pm4py.algo.analysis.woflan import algorithm as woflan
import os
import pandas as pd
import numpy as np
```

```python
logs_metadir = "C:\\Users\\matte\\Desktop\\results\\synthetic\\classic_miners"
```

# Utils

```python
def discover_with_alpha(log, filename):
    net, im, fm = pm4py.discover_petri_net_alpha(log)
    fitness = pm4py.fitness_token_based_replay(log, net, im, fm)
    prec = pm4py.precision_token_based_replay(log, net, im, fm)
    gen = generalization_evaluator.apply(log, net, im, fm)
    simp = simplicity_evaluator.apply(net)
    is_sound = woflan.apply(net, im, fm, parameters={woflan.Parameters.RETURN_ASAP_WHEN_NOT_SOUND: True,
                                                     woflan.Parameters.PRINT_DIAGNOSTICS: False,
                                                     woflan.Parameters.RETURN_DIAGNOSTICS: False})
    pm4py.write_pnml(net, im, fm, filename+".pnml")
    pm4py.save_vis_petri_net(net, im, fm, filename+"net.png")
    return net, im, fm, fitness, prec, gen, simp, is_sound
```

# Data

```python
df_fitness = []
df_precision = []
df_generalization = []
df_simplicity = []
df_soundness = []
df_fscore = []

datasets = []
folders = os.listdir(logs_metadir)
for folder in folders:
    log_data_dir = os.path.join(logs_metadir, folder)
    logs = [file for file in os.listdir(log_data_dir) if "xes" in file]

    alpha_dir = os.path.join(log_data_dir, "alpha")
    alpha_plus_dir = os.path.join(log_data_dir, "alpha_plus")
    heuristics_dir = os.path.join(log_data_dir, "heuristics")
    inductive_dir = os.path.join(log_data_dir, "inductive")

    dirs = [alpha_dir, alpha_plus_dir, heuristics_dir, inductive_dir]
    for dir in dirs:
        if not os.path.exists(dir):
            os.mkdir(dir)

    for log_name in logs:
        filename = os.path.join(log_data_dir, log_name)
        log = pm4py.read_xes(filename)

        # alpha
        partial_filename = os.path.join(alpha_dir, log_name.split('.')[0])
        _, _, _, fitness, precision, generalization, simplicity, soundness = discover_with_alpha(log, partial_filename)

        res_string = f"fitness: {fitness}\nprecision: {precision}\ngeneralization: {generalization}\nsimplicity: {simplicity}\nsoundnedd: {soundness}"

        with open(os.path.join(alpha_dir, log_name.split('.')[0]+"res.txt"), "w") as file:
            file.write(res_string)


        # alpha plus
        partial_filename = os.path.join(alpha_plus_dir, log_name.split('.')[0])
        _, _, _, fitness, precision, generalization, simplicity, soundness = discover_with_alpha_plus(log, partial_filename)

        res_string = f"fitness: {fitness}\nprecision: {precision}\ngeneralization: {generalization}\nsimplicity: {simplicity}\nsoundnedd: {soundness}"

        with open(os.path.join(alpha_plus_dir, log_name.split('.')[0]+"res.txt"), "w") as file:
            file.write(res_string)


        # heuristics
        partial_filename = os.path.join(heuristics_dir, log_name.split('.')[0])
        _, _, _, fitness, precision, generalization, simplicity, soundness = discover_with_heuristics(log, partial_filename)

        res_string = f"fitness: {fitness}\nprecision: {precision}\ngeneralization: {generalization}\nsimplicity: {simplicity}\nsoundnedd: {soundness}"

        with open(os.path.join(heuristics_dir, log_name.split('.')[0]+"res.txt"), "w") as file:
            file.write(res_string)

        # inductive
        partial_filename = os.path.join(inductive_dir, log_name.split('.')[0])
        _, _, _, fitness, precision, generalization, simplicity, soundness = discover_with_inductive(log, partial_filename)

        res_string = f"fitness: {fitness}\nprecision: {precision}\ngeneralization: {generalization}\nsimplicity: {simplicity}\nsoundnedd: {soundness}"

        with open(os.path.join(inductive_dir, log_name.split('.')[0]+"res.txt"), "w") as file:
            file.write(res_string)

    
```
