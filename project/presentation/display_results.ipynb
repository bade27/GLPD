```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib import rcParams

%matplotlib inline

# figure size in inches optional
rcParams['figure.figsize'] = 20 ,15

discovered_dir = "C:\\Users\\matte\\Desktop\\dataset_union_original\\data\\test_graphs\\inference\\images\\"
input_dir = "C:\\Users\\matte\\Desktop\\dataset_union_original\\data\\test_graphs\\saved_images_pnml\\input_net\\images\\"

discovered_nets = [os.path.join(discovered_dir, name) for name in sorted(os.listdir(discovered_dir))]
input_nets = [os.path.join(input_dir, name) for name in sorted(os.listdir(input_dir))]

discovered_iter = iter(discovered_nets)
input_iter = iter(input_nets)
```

```python
next_discovered = next(discovered_iter)
next_input = next(input_iter)

img_A = mpimg.imread(next_discovered)
img_B = mpimg.imread(next_input)

# display images
fig, ax = plt.subplots(1,2)
ax[0].imshow(img_A)
ax[1].imshow(img_B)
```
